### 经典框架

- LeNet
   
   1. 网络结构
   
   
   2conv + 3fc

- AlexNet（2012）
   1. 网络结构
   
   
   kernel size: 11x11
   2. 论文创新点
      1. 使用了两块3G显存GTX580的显卡训练一周（速度提升）
      2. 网络层次更深了(60M params 5conv+3fc layers)
      3. 使用了Dropout试图解决overfitting问题
      4. Relu
      5. Max Pooling
   3. 缺点
      1. 卷积核尺寸过大
   
   

   
- VGG（2014）
   
   1. 网络结构
   
   
   2. 论文创新点
     1. 卷积核尺寸3x3、1x1

- GoogLeNet（2014）
   1. 网络结构
   
   > 22层网络

<img src = "img/googlenet 02.png">



多个输出的意思是有多个版本的GoogleNet

   
   2. 论文创新点
      1. 同一层有多个不同类型的卷积核
      
      > 多个不同大小的卷积核做卷积后concat
      
      <img src = "img/googlenet 01.png">
      2. 简单的堆叠网络层数堆叠到22层左右再叠加层数不会使模型更优，原因是层数过深、反向传播时输入层端的layer出现梯度消失

- ResNet（2016）

   1. 创新点
      1. ShortCut（短路）
      <img src="./img/resnet 01.png"> 
      
   2. 网络结构
      1. Unit 三个Conv+BN + 短路
      2. 多个Unit串联（常见152 layers）
