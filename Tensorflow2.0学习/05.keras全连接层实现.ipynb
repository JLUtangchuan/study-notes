````python
import tensorflow as tf

x = tf.random.normal((10,5))
````

````python
layer1 = tf.keras.layers.Dense(3)  # 可选参数 activation

out = layer1(x)
````

````python
layer1.bias  # bias 初始化是0
````

````python
layer1.kernel # weight初始化好像是truncked_normal
````

#### 实现嵌套的多层网络
`
model = keras.Sequential([layer1,layer2,layer3])
model.build(input_shape)
model.summary
`
-  model.summary函数
    输出网络结构的函数


````python
from tensorflow import keras

x = tf.random.normal([5,2])
h1 = keras.layers.Dense(3,activation='relu')
h2 = keras.layers.Dense(3,activation='relu')
o = keras.layers.Dense(2,activation='sigmoid')

model = keras.Sequential([h1,h2,o])
model.build(input_shape=x.shape)
model.summary()
````

````python
type(model) #tensorflow.python.keras.engine.sequential.Sequential
````
