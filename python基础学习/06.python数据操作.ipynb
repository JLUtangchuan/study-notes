### **MAP REDUCE**介绍

map reduce是用来处理大规模数据的工具


*原因见Google的[“MapReduce: Simplified Data Processing on Large Clusters”](http://research.google.com/archive/mapreduce.html)*

---

map： map(f,list)

   返回f(list)
    
   通常f是一个lambda表达式

---

reduce:

   把一个函数作用在一个序列 [x1, x2, x3...]上，这个函数必须接收两个参数，reduce把结果继续和序列的下一个元素做累积计算

`
reduce(f, [x1, x2, x3, x4]) = f(f(f(x1, x2), x3), x4)
`
    
> python3中reduce函数被取消了，放入到了functools模块中
---

filter:
    过滤掉不符合条件的元素（符合条件的设置为一组）
    
        filter(function, iterable)
        
---

***注意：上述方法在python3中返回的不是列表而是对应的对象，可以转成list or tuple***

````python
# map reduce
arr = map(lambda x:x**2,range(4))
print(list(arr))

from functools import reduce
re = reduce(lambda a1,a2:a2+a1,range(5))
print(re)
````

### Python 内置库函数

*此类库函数方便我们对数据进行处理*

---

- zip 函数
   - 是一个类
   - 方便我们将两个数组对应元素组合成一个tuple，从而形成一个新列表
   - 例：机器学习中打包(X,y)
   - 长度不一致时，返回最短长度
   ```
   c = list(zip(a,b))
   
   ```
- enumerate 函数
   - 返回一个新数组  等价于 zip(range(len(seq)),seq)
   - 可以在for in 遍历中使用
   - 默认参数 start=0 设置起始下标号
   ```
   for index,element in enumerate(seq):
       pass
   
   ```
---

- collections 模块
   - 提供了操作集合的类
   - namedtuple 建立一个有名字的实体类
      - NewClass = collections.namedtuple(typename,field_names)
          
          field_names是一个str列表
          
          
   - defaultdict 查找Key不存在时，返回默认值   
      - defdict = collections.defaultdict(lambda:'N/A')
      - 用途： 在词频统计时，当查找到0次出现的词（未在字典中出现）不会报错
      
   - ***Counter 统计次数***
      - 返回一个字典，记录列表中重复元素的次数

---

- itertools 模块
   - 提供了用于操作迭代对象的函数
   - chain 将多个迭代器串联
   - groupby 将***相邻的***重复元素放在一起形成dict（Key:[Key,Key,Key]）
      - 注意是相邻元素，如果不相邻，选择sort
      - 默认参数key=None 接收一个函数（hash）用于判断是否是同一个group
      

````python
a = [3,5,6,'sda']

print(list(enumerate(a)))
````

````python
# 实现文字标记转标号（建立映射）
def str2idx(li):
    s = set(li)
    return list(enumerate(s))

# 加入统计次数功能
def str2idx_plus(li):
    s = set(li)
    def f(i):
        return (i,li.count(i))
    s = list(map(f,s))
    s = sorted(s,key = lambda i: i[-1],reverse=True)
    return list(enumerate(s,start = 1))

l = ['a','b','a','c','a','b']
u = str2idx_plus(l)
print(u)
````
