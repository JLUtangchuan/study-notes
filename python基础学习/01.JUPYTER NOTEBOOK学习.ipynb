快捷键
    . ESC 退出编辑、此时上下移动的是cell位置(进入命令模式)
    . ENTER 进入编辑
    
    命令模式下的操作
   . A 上方插入cell
   . B 下方插入cell
   . DD 删除单元格
   . Z 撤销删除
   . Y M code or md
   . Shift+up 多选
   . Shift+M 合并cell
   . F 查找
   . 1 2 3 heading（for md）
   
    编辑模式下的操作
   . ctrl + s 保存
   . ctrl + enter 运行单元格
   . alt + enter 运行单元格并加入新单元格
    
    
   ctrl + shift + f 代码命令选项板
