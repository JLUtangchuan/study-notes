```python
import pandas as pd
import numpy as np

# 构建Series

a  =  pd.Series(np.arange(3),index=['a','b','c'])

print(a.dtype)
print(a.index)
a['b']

'b' in a

dic = {'a':'sb','b':'sda','c':1000}
b = pd.Series(dic)

b['a'] = 12
b['d'] = 1000

b
print(b['d'])

```

```python
# 构建DataFrame

df = pd.DataFrame(np.arange(10).reshape(2,5),index = ['a','b'],columns=[4,5,6,7,8])

# 基本属性
#
print(df.columns)
print(df.index)
print(type(df.values))

print(df.shape)
print(df.dtypes)


# 常见操作
print(df.describe())

print(df.head(1)) # 看前1行的数据
print(df.tail(1))

print(df.sample(n = 1)) # 随机取3行



#  
df.sort_index(axis=1)
df

df = pd.DataFrame(pd.read_csv('./dataset/cleveland.csv',header = None))

df = df.head(10)
df.shape

# 删除某行
df.drop(9,axis=0)
# 删除列
df.drop(0,axis=1)



# 索引切片

#loc 函数 location  类似于list的slice 通过label索引
print(df.loc[3:5,(3,5,7)]) # 行 + 列

#iloc index location 通过行列号索引

# 访问单个元素 at
print(df.at[5,5])

# 访问某一列
print(df[5])


# 计算
# max、min、mean、std、count、median，cumsum、sum、diff
print(df[3][5])  # 3列5行
# 补全缺失值
df.fillna(value = 'error')  # method参数 ffill bfill  （forword fill

df
```
