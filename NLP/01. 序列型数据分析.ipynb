2. [words,word vec]
> 不只是逻辑上的转换，需要语义上的转换
*方法*
   - one hot encoding
      特点
      - sparse（开销）
      - high dim
      - 忽略了词与词之间的联系（语义相关性）
   - word embedding
      - 把每一个单词编码成一个维度略低向量（向量内容可机器学习获得）
      - word2vec
      - GloVe
